```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split #library to split dataset into training and testing
from sklearn.metrics import accuracy_score

music_data = pd.read_csv('music.csv')
#need to split daya into input and output set
#drop method creates new dataset without specified column
X= music_data.drop(columns= ('genre'))

#output data
y= music_data['genre']

#splitting data and allocatin 20% for testing_done randomly)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

#create new instance
model = DecisionTreeClassifier()

#Training model to learn pattern
model.fit(X_train,y_train)

#ask model to make predictions with test data

predictions = model.predict(X_test)
predictions

#how to measure accuracy: compare predictions with output test data
score = accuracy_score(y_test, predictions)
score
```
